```python
import youtube_dl

def get_youtube_info(url: str):
    ydl_opts = {}
    with youtube_dl.YoutubeDL(ydl_opts) as ydl:
        info_dict = ydl.extract_info(url, download=False)
        title = info_dict.get('title', None)
        desc = info_dict.get('description', None)
    return title, desc
```

```python
from pytubefix import YouTube
import time

def get_youtube_info(url, max_retries=5, interval_secs=5):
    for attempt in range(max_retries):
        try:
            yt = YouTube(url)
            title = yt.title
            description = yt.description
            return title, description
        except KeyError:
            print(f"Attempt {attempt + 1} failed. Retrying in {interval_secs} seconds...")
            time.sleep(interval_secs)
    raise Exception("Failed to retrieve video details after several attempts")

url = "https://youtube.com/watch?v=bvPDQ4-0LAQ"
title, description = get_youtube_info(url)
print(f"Title: {title}\nDescription: {description}")
```

```python
import requests
from bs4 import BeautifulSoup

def get_youtube_description(url: str) -> str:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    description_tag = soup.find('meta', {'name': 'description'})
    if description_tag:
        return description_tag['content']
    return "Description not found"

url = "https://youtube.com/watch?v=bvPDQ4-0LAQ"
description = get_youtube_description(url)
print(f"Description: {description}")
```

```python
import yt_dlp

def get_youtube_transcript(url: str):
    ydl_opts = {}
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        info_dict = ydl.extract_info(url, download=False)
        captions = info_dict.get('automatic_captions', {}).get('en', [])
        print("Captions structure:", captions)  # Debugging line
        if captions:
            for caption in captions:
                print(caption)  # Print each caption to inspect its structure
        transcript = [{'text': caption['text']} for caption in captions]
        return transcript

# Example usage
url = "https://youtube.com/watch?v=bvPDQ4-0LAQ"
transcript = get_youtube_transcript(url)
print(transcript)
```

```python
from langchain.document_loaders import YoutubeLoader
from pytube import YouTube
from pytube.exceptions import PytubeError

def get_youtube_transcript_loader_langchain(url: str):
    try:
        # Initialize the YoutubeLoader with the correct URL
        loader = YoutubeLoader.from_youtube_url(url, add_video_info=True)
        return loader.load()
    except KeyError as e:
        print(f"KeyError: {e} - The video details could not be accessed.")
    except PytubeError as e:
        print(f"PytubeError: {e} - There was an issue with pytube.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

def wrap_docs_to_string(docs):
    return " ".join([doc.page_content for doc in docs]).strip()

def process_youtube_transcript(url: str):
    # Load the transcript from the YouTube video
    docs = get_youtube_transcript_loader_langchain(url)
    
    if docs:
        # Convert the document contents into a single string
        transcript_string = wrap_docs_to_string(docs)
        return transcript_string
    else:
        return "Failed to load transcript."

# Example usage
youtube_url = "https://www.youtube.com/watch?v=example_video"
transcript = process_youtube_transcript(youtube_url)
print(transcript)
```

```python
import yt_dlp
import os
import glob

def download_transcripts(video_urls, download_path):
    """
    Downloads transcripts for a list of YouTube video URLs.
    
    :param video_urls: List of YouTube video URLs.
    :param download_path: Path to the directory where transcripts will be saved.
    """
    ydl_opts = {
        'skip_download': True,
        'writesubtitles': True,
        'allsubtitles': True,  # Download all available subtitles
        'subtitleslangs': ['en'],  # Specify the language you want
        'outtmpl': os.path.join(download_path, '%(title)s.%(ext)s')
    }
    
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download(video_urls)

def process_transcripts(directory):
    """
    Processes YouTube video transcripts stored in text files within a specified directory.
    Reads the contents of each file and concatenates them into a single string.
    
    :param directory: The directory containing the transcript files.
    :return: A single string containing the concatenated contents of all transcript files.
    """
    all_contents = []
    
    # Use glob to find all .vtt files in the directory
    for filepath in glob.glob(f"{directory}/*.vtt"):
        with open(filepath, 'r', encoding='utf-8') as file:
            content = file.read()
            all_contents.append(content)
    
    # Concatenate all contents into a single string
    combined_content = "\n".join(all_contents)
    
    return combined_content

# Example usage
video_urls = [
    'https://www.youtube.com/watch?v=NMObj5tBKX4&t=3748s'
]
download_path = "transcripts"

# Create the directory if it doesn't exist
os.makedirs(download_path, exist_ok=True)

# Download transcripts
download_transcripts(video_urls, download_path)

# Process transcripts
combined_transcripts = process_transcripts(download_path)
print(combined_transcripts)
```

```python
import tiktoken

def wrap_docs_to_string(docs):
    return " ".join([doc.page_content for doc in docs]).strip()

def get_youtube_transcript_loader_langchain(url):
    # This is a placeholder for the actual LangChain YouTube loader function
    # Replace this with the actual implementation
    from langchain.document_loaders import YoutubeLoader
    loader = YoutubeLoader.from_youtube_url(url)
    return loader.load()

def get_youtube_transcription(url: str):
    try:
        # Load the transcript using LangChain's YouTube loader
        docs = get_youtube_transcript_loader_langchain(url)
        
        # Combine the document contents into a single string
        text = wrap_docs_to_string(docs)
        
        # Encode the text using the tiktoken encoder for the GPT-4 model
        enc = tiktoken.encoding_for_model("gpt-4")
        count = len(enc.encode(text))
        
        return text, count
    except Exception as e:
        # Return an error message and a count of 0 in case of an exception
        return f"Error: {e}", 0

# Example usage
url = 'https://www.youtube.com/watch?v=3XTOtqSp7Mw'
transcription, token_count = get_youtube_transcription(url)
print(f"Transcription: {transcription}")
print(f"Token Count: {token_count}")
```

```python
import yt_dlp

def list_subtitles(video_url):
    """
    Lists all available subtitles for a YouTube video URL.
    
    :param video_url: YouTube video URL.
    """
    ydl_opts = {
        'listsubtitles': True
    }
    
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        info_dict = ydl.extract_info(video_url, download=False)
        subtitles = info_dict.get('subtitles', {})
        for lang, subs in subtitles.items():
            print(f"Language: {lang}")
            for sub in subs:
                print(f"  - {sub['ext']}")

# Example usage
video_url = 'https://www.youtube.com/watch?v=NMObj5tBKX4&t=3748s'
list_subtitles(video_url)
```
